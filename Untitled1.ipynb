```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
try:
    # Configurar opções do Chrome para headless mode
    chrome_options = Options()
    # chrome_options.add_argument("--headless")  # Rodar sem interface gráfica
    # chrome_options.add_argument("--disable-gpu")  # Necessário para algumas versões do Windows
    # chrome_options.add_argument("--no-sandbox")  # Necessário para Linux
    # chrome_options.add_argument("--disable-dev-shm-usage")  # Necessário para contornar problemas de memória em containers
    
    # Inicializar o WebDriver com as opções configuradas
    driver = webdriver.Chrome(options=chrome_options)
    
    # Abrir a página no navegador
    driver.get('https://br.investing.com/equities/petrobras-pn')
    
    # Esperar o carregamento da página, se necessário (opcional)
    driver.implicitly_wait(10)
    
    # Pegar o HTML renderizado
    html = driver.page_source
finally:
    driver.quit()

# Utilizar BeautifulSoup para processar o HTML
soup = BeautifulSoup(html, 'html.parser')

# Exemplo de uso: imprimir o título da página
print(soup.title.string)
```
